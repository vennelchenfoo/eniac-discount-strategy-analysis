```python
import pandas as pd

orders_avg_diff_paid_no_outliers = pd.read_csv("../data/quality-assessed-data/orders_avg_diff_paid_no_outliers.csv", dtype=str)
product_category_df = pd.read_csv("../data/categorized-data/product_category_df.csv")
orderlines_qu = pd.read_csv("../data/quality-assessed-data/orderlines_qu.csv", dtype=str)
```

```python
orderlines_order_no_outliers = orders_avg_diff_paid_no_outliers.merge(
    orderlines_qu,
    on='order_id',
    how='left'
)
orderlines_order_no_outliers
```

```python
final_joined_df = orderlines_order_no_outliers.merge(product_category_df, how="left" , on="sku")
```

## Here we are calculating the discount and the discount in percentage for each item in orders

```python
final_joined_df["price"] = pd.to_numeric(final_joined_df["price"])
final_joined_df["unit_price"] = pd.to_numeric(final_joined_df["unit_price"])

final_joined_df['discount'] = (final_joined_df['price'] - final_joined_df['unit_price']).round(2)
final_joined_df['discount_pct'] = (final_joined_df['discount'] / final_joined_df['price'] * 100).round(2)
```

### Categorize based on discount in different ranges

```python
# Define discount ranges for analysis
def categorize_discount(discount_pct):
    if discount_pct <= 0:
        return '0 - No Discount'
    elif discount_pct <= 15:
        return '1-15%'
    elif discount_pct <= 25:
        return '16-25%'
    elif discount_pct <= 45:
        return '26-45%'
    elif discount_pct <= 55:
        return '46-55%'
    elif discount_pct <= 65:
        return '56-65%'
    elif discount_pct <= 75:
        return '66-75%'
    else:
        return '75%+'

# Apply the categorization function to the final_joined_df
final_joined_df['discount_range'] = final_joined_df['discount_pct'].apply(categorize_discount)
```

### Categorize based on discount in different ranges, more granularity

```python
# Define discount ranges for analysis
def categorize_discount_more_granularity(discount_pct):
    if discount_pct <= 0:
        return '0 - No Discount'
    elif discount_pct <= 5:
        return "1-5%"
    elif discount_pct <= 10:
        return "6-10%"
    elif discount_pct <= 15:
        return "11-15%"
    elif discount_pct <= 20:
        return "16-20%"
    elif discount_pct <= 25:
        return "21-25%"
    elif discount_pct <= 30:
        return "26-30%"
    elif discount_pct <= 35:
        return "31-35%"
    elif discount_pct <= 40:
        return "36-40%"
    elif discount_pct <= 45:
        return "41-45%"
    elif discount_pct <= 50:
        return "46-50%"
    elif discount_pct <= 55:
        return "51-55%"
    elif discount_pct <= 60:
        return "56-60%"
    elif discount_pct <= 65:
        return "61-65%"
    elif discount_pct <= 70:
        return "66-70%"
    elif discount_pct <= 75:
        return "71-75%"
    elif discount_pct <= 80:
        return "76-80%"
    elif discount_pct <= 85:
        return "81-85%"
    elif discount_pct <= 90:
        return "86-90%"
    elif discount_pct <= 95:
        return "91-95%"
    else:
        return "96%+"


# Apply the categorization function to the final_joined_df
final_joined_df['discount_range_more_granularity'] = final_joined_df['discount_pct'].apply(categorize_discount_more_granularity)
```
